```python
import math # Importar el módulo math para usar math.pi
#solicitar la eleccion del usuario
opcion = input("que figura quiere elegir? (escriba t para triangulo y c para circulo)")
opcion = opcion.upper() #convertir las mayusculas para manejar ´t´ o ´c´ minusculas

#estructura condicional de calculo
if opcion == "T":
    # --- calculo del area del triangulo ---
    print("\n--- Cálculo del Triángulo ---")
    try:
        base = float(input("Ingrese la base del triángulo: "))
        altura = float(input("Ingrese la altura del triángulo: "))
        area = (base * altura) / 2
        print("El área del triángulo es:", area)
    except ValueError:
        print("Entrada no válida. Por favor, introduce un número.")
elif opcion == 'C':
    # --- Cálculo del Área del Círculo ---
    print("\n--- Cálculo del Círculo ---")
    try:
        radio = float(input("Ingrese el radio del círculo: "))

        # Fórmula: PI * radio * radio
        # Usamos math.pi para mayor precisión
        area = math.pi * (radio ** 2)
        print(f"\nEl área del Círculo es: {area:.2f} unidades cuadradas.")

    except ValueError:
        print("Error: Por favor, ingrese un valor numérico válido para el radio.")

else:
    # --- Manejo de opción no válida ---
    print("\nOpción no válida. Por favor, escriba 'T' o 'C'.")
```

```python
def calcular_area_rectangulo():
  """
  Calcula el área de un rectángulo
  """
  print("\n--- Calculadora del Rectángulo ---")
  try:
    lado1 = float(input("Ingrese la longitud del primer lado (base): "))
    lado2 = float(input("Ingrese la longitud del segundo lado (altura): "))

    if lado1 > 0 and lado2 > 0: # Corrected condition to check for positive sides
      area = lado1 * lado2
      print(f"\nAmbos lados son positivos: {lado1} y {lado2}")
      print(f"El área del rectángulo es: {lado1} * {lado2} = {area}")
    else:
      print("\nError de entrada:")
      print("Ambos lados deben ser valores positivos y mayores que cero para calcular el área.")
      print(f"Lados ingresados: {lado1} y {lado2}")
  except ValueError:
    print("Error: Por favor, ingrese valores numéricos válidos para los lados del rectángulo.")

# Llamar a la función para ejecutarla
calcular_area_rectangulo()
```

```python
def calcular_pago_restaurante(consumo_total):
    """
    Calcula el pago final de un consumo de restaurante, aplicando un descuento
    del 15% si el monto excede los $130.000.
    """
    UMBRAL_DESCUENTO = 130000.00
    DESCUENTO_PORCENTAJE = 0.15

    descuento_monto = 0.0
    pago_final = consumo_total

    print(f"--- Cálculo de Pago ---")
    print(f"Consumo Total Inicial: ${consumo_total:,.2f}")

    # Evaluar la condición del descuento
    if consumo_total > UMBRAL_DESCUENTO:
        # Aplicar Descuento
        descuento_monto = consumo_total * DESCUENTO_PORCENTAJE
        pago_final = consumo_total - descuento_monto

        print("\n ¡Descuento Aplicado!")
        print(f"Se ha superado el umbral de ${UMBRAL_DESCUENTO:,.2f}.")
        print(f"Descuento (15%): ${descuento_monto:,.2f}")
    else:
        # No hay Descuento
        print("\n Sin Descuento")
        print(f"El consumo no excede el umbral de ${UMBRAL_DESCUENTO:,.2f}.")

    print("\n-------------------------")
    print(f"Monto Final a Pagar: **${pago_final:,.2f}**")
    print("-------------------------")


calcular_pago_restaurante(999)
```

```python
def calcular_pares_impares_if_elif():
  suma_pares = 0
  producto_impares = 1
  print("--- Suma de Pares y Producto de Impares ---")
  try:
    N = int(input("¿Cuántos números desea ingresar (N)? "))
  except ValueError:
    print("Error: Ingrese un número entero válido para N.")
    return

  for i in range(1, N + 1):
    try:
      numero_actual = int(input(f"Ingrese el número {i}: "))
      if numero_actual % 2 == 0:
        suma_pares += numero_actual
        print(f"  -> {numero_actual} es PAR. Suma actual: {suma_pares}")
      elif numero_actual % 2 != 0:
        producto_impares *= numero_actual
        print(f"  -> {numero_actual} es IMPAR. Producto actual: {producto_impares}")
    except ValueError:
      print("Advertencia: Entrada inválida. Se omitió este valor.")
      continue

  print("\n--- Resultados ---")
  print(f"La suma de los números pares ingresados es: {suma_pares}")
  print(f"El producto de los números impares ingresados es: {producto_impares}")


```
